```python

import pandas as pd #importar pandas
import pandera as pa #importar pandera
```

```python
df = pd.read_csv("ocorrencia.csv",encoding='latin-1',sep=';',parse_dates=['ocorrencia_dia'],dayfirst=True)#le arquivo CSV mudar o encoding para latin e o separador para ; e converte campos para data e colocando o dia primeiro.
df.head(10)#imprimir as 10 primeiras linhas 
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String), 
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String,pa.Check.str_length(2,2)),#limitar os caracteres em 2 digitios 
        "ocorrencia_aerodromo":pa.Column(pa.String),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String,pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'),nullable=True),#permite null
        "total_recomendacoes":pa.Column(pa.Int)
        
        
    }
)
```

```python
schema.validate(df)
```
